```python
import pandas as pd

def save_dataframe_investing_cards(threshold=15):
    # Lire le fichier CSV original
    df = pd.read_csv('pokemon_cards.csv')
    
    # Crer le masque pour filtrer les cartes > threshold
    mask = (df['holofoil_price'] > threshold) | (df['reverse_holofoil_price'] > threshold)
    
    # Filtrer et trier par prix dcroissant
    cards_over_threshold = df[mask].sort_values(by=['holofoil_price', 'reverse_holofoil_price'], ascending=False)
    
    # Crer le nouveau nom de fichier en ajoutant '_15'
    new_filename = f'pokemon_cards_{threshold}.csv'
    
    # Sauvegarder le nouveau fichier CSV
    cards_over_threshold.to_csv(new_filename, index=False, encoding='utf-8', float_format='%.2f')
```

```python
save_dataframe_investing_cards()
```

```python
import os
import pandas as pd
import time
from tqdm import tqdm

def sequential_price_extraction(cards_df, output_dir='price_history'):
    os.makedirs(output_dir, exist_ok=True)
    
    progress_bar = tqdm(total=len(cards_df), desc="Extraction des prix")
    
    for index, card_row in cards_df.iterrows():
        card_id = card_row['id']
        file_path = os.path.join(output_dir, f'{card_id}.csv')
        
        if os.path.exists(file_path):
            progress_bar.set_description(f"Fichier existant pour {card_id}")
            progress_bar.update(1)
            continue
            
        try:
            price_history = extract_price_history(f"https://prices.pokemontcg.io/tcgplayer/{card_id}")
            
            # Rinitialisation de l'index pour avoir les dates en colonnes
            price_df = price_history.reset_index()
            
            # Renommage des colonnes pour correspondre au format souhait
            price_df.columns = ['start_date', 'end_date', 'price', 'quantity_sold']
            
            # Sauvegarde au format CSV
            price_df.to_csv(file_path, index=False)
            
            progress_bar.set_description(f"Extraction russie pour {card_id}")
            
        except Exception as e:
            progress_bar.set_description(f"chec pour {card_id}: {str(e)}")
        
        progress_bar.update(1)
        time.sleep(1)
    
    progress_bar.close()
```

```python
sequential_price_extraction(ts_cards.tail(3))
```
